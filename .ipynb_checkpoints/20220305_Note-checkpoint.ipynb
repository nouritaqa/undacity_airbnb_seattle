```python
def dollar_float(x):
    if type(x)==str:
        x=x[1:].replace(",","")
        x=float(x)
    return x
```

```python
def perc_float(x):
    if type(x)==str:
        x=x.replace('%','')
        x=float(x)/100.0
    return (x)
```

```python
#Covert available to numeric data
df_cal['available'] = df_cal['available'].apply(lambda x:1 if x =='t' else 0)
```

```python
#Create Year, Month, Day, Day Name columns after convering date into datetime
df_cal['date'] = pd.to_datetime(df_cal['date'],format="%Y-%m-%d")
df_cal['year'] = df_cal['date'].dt.year
df_cal['month'] = df_cal['date'].dt.month
df_cal['day'] = df_cal['date'].dt.day
df_cal['day_name'] = df_cal['date'].dt.day_name()
df_cal['day_num'] = df_cal['date'].dt.weekday
df_cal.head()
```

```python
# plot average price and total listing to see seasonal patterns
fig,ax = plt.subplots()
ax.plot(df_cal.groupby(['month']).price.mean(),color='slategrey', marker='o', linewidth=3, label='Price')
ax.set_xlabel('Month')
ax.set_ylabel('Average Price, $')

ax2 = ax.twinx()
ax2.plot(df_cal.groupby(['month']).available.sum(), color='orange', marker ='o', linewidth=3, label='Listing')
ax2.set_ylabel('Total Available Listings')

ax.legend(loc=4)
ax2.legend(loc=2)
plt.show()
```

```python
# plot average price and total listing to see seasonal patterns
fig,ax = plt.subplots()
ax.plot(df_cal.groupby(['day_num']).price.mean(),color='slategrey', marker='o', linewidth=3, label='Price')
ax.set_xlabel('Day')
ax.set_ylabel('Average Price, $')

ax2 = ax.twinx()
ax2.plot(df_cal.groupby(['day_num']).available.sum(), color='orange', marker ='o', linewidth=3, label="Listing")
ax2.set_ylabel('Total Available Listings')

ax.legend(loc=4)
ax2.legend(loc=2)
plt.show()
```

Findings
- It is clear that average price is highest in July, probably due to the summer holiday season.
- Total number of listings is lowest in January and highetst in December. It seems there is no clear relationship between the toal number of listings and average price from this graph. Number of listing might be affected by different factors.
- There is a sharp drop in both price and # of listing between December and January, implying the high demand winter holidays just before new year, and rapid decrease during January. 
- At weekly basis, the price is highet during the weekend from Fri to Sat. On the contrary, # of listing increases from Sun to Monday, also indicating non-price factors that affect the number of listing.
- Taken together, to travel low-cost, it is advisoable to avoid holiday season as well as Friday and Saturday.

```python
cols =['neighbourhood_cleansed',"neighbourhood_group_cleansed",
       'property_type','city','smart_location','bed_type','host_response_time',
       'cancellation_policy','room_type']
```

```python
df_list[cols]
```

```python
df_list[['price','summary']].head()
```

```python
df_list[['price','accommodates']]
```

```python
df_list_rf = pd.concat([df_list[cols],df_list[['price','bedrooms']]],axis=1)
df_list_rf
```

```python
df_list_rf.bedrooms.isnull().sum()
```

```python
df_list_rf.dropna(subset=['bedrooms'],inplace=True)
```

```python
df_list_rf.bedrooms.isnull().sum()
```

```python
obj_col = df_list[cols].columns
obj_col
```

```python
for col in obj_col:
    df_list_rf = pd.concat([df_list_rf.drop(col,axis=1),pd.get_dummies(df_list_rf[col],prefix=col,prefix_sep='_',drop_first=True, dummy_na=True)],axis=1)
```

```python
df_list_rf.head()
```

```python
X = df_list_rf.drop(columns='price')
y = df_list_rf.price
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
rforest = RandomForestClassifier(n_estimators=100,n_jobs=-1,)
```

```python
rforest.fit(X_train,y_train)
```

```python
y_train_preds = rforest.predict(X_train)
y_test_preds = rforest.predict(X_test)

print('Random Forest MSE train: %.3f, test: %.3f' % (
        mean_squared_error(y_train, y_train_preds),
        mean_squared_error(y_test, y_test_preds)))
print('Random Forest R^2 train: %.3f, test: %.3f' % (
        r2_score(y_train, y_train_preds),
        r2_score(y_test, y_test_preds)))
```
