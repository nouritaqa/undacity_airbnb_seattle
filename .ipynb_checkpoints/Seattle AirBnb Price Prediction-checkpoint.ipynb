```python
#Reading Data
df_list = pd.read_csv('listings.csv')
df_cal = pd.read_csv('calendar.csv')
df_rev = pd.read_csv('reviews.csv')
```

```python
#Functions for Data Preparation
def dollar_float(x):
    '''
    input: dollar value with string dtype
    output: value in float
    '''
    if type(x)==str:
        x=x[1:].replace(",","")
        x=float(x)
    return x

def perc_float(x):
    '''
    input: % value with string dtype
    output: % value in float
    '''
    if type(x)==str:
        x=x.replace('%','')
        x=float(x)/100.0
    return (x)
```

## Q1.Is there any seasonal variability in prices and room availability?

```python
df_cal.info()
```

```python
df_cal.head()
```

- price should be converted to float
- date will be seprated to year/month/day

#### Data Preparation

```python
#Removes rows missing price data since we are interested in price vairability
df_cal.dropna(subset=['price'],inplace=True)

#Convert dollar dtype to float
df_cal['price']=df_cal['price'].apply(lambda x: dollar_float(x))

#Create Year, Month, Day, Day Name columns after convering date into datetime
df_cal['date'] = pd.to_datetime(df_cal['date'],format="%Y-%m-%d")
df_cal['year'] = df_cal['date'].dt.year
df_cal['month'] = df_cal['date'].dt.month
df_cal['day'] = df_cal['date'].dt.day
df_cal['day_name'] = df_cal['date'].dt.day_name()
df_cal['day_num'] = df_cal['date'].dt.weekday

#Covert available to numeric data
df_cal['available'] = df_cal['available'].apply(lambda x:1 if x =='t' else 0)

df_cal.head()
```

```python
df_cal.describe()
```

- All rows with price data is available "true' status (now it's converted to 1), indicating that we can consider these are listing for reservation for the given day.

```python
df_cal[df_cal.year==2017].groupby('month').available.sum()
```

For the rest of analysis, since we want to compare prices and # of listings between months, we remove the 2017's data, which is only for January.

```python
# Plot average price and total number of listings per month to see seasonal patterns
fig,ax = plt.subplots()
ax.plot(df_cal[df_cal.year==2016].groupby(['month']).price.mean(),color='blue', marker='o', linewidth=3, label='Price')
ax.set_xlabel('Month')
ax.set_ylabel('Average Price, $')

ax2 = ax.twinx()
ax2.plot(df_cal[df_cal.year==2016].groupby(['month']).available.sum(), color='orange', marker ='o', linewidth=3, label='Listing')
ax2.set_ylabel('# of Available Listings')

ax.legend(loc=2)
ax2.legend(loc=4)

plt.title('Seasonal Change of Price and # of Available Listings')
plt.show()
```

```python
# Plot average price and total listing to weekly patterns
fig,ax = plt.subplots()
ax.plot(df_cal.groupby(['day_num']).price.mean(),color='slategrey', marker='o', linewidth=3, label='Price')
ax.set_xlabel('Day')
ax.set_ylabel('Average Price, $')

ax2 = ax.twinx()
ax2.plot(df_cal.groupby(['day_num']).available.sum(), color='orange', marker ='o', linewidth=3, label="Listing")
ax2.set_ylabel('Total Available Listings')

ax.legend(loc=2)
ax2.legend(loc=4)

plt.title('Weekly Change of Price and # of Available Listings')
plt.show()
```

### Findings and answers to the Question 1:
- Regarind the average price, it is the highest in July for monthly basis and on Friday (4 on X-axis) and Saturday (5) for weekly basis. This implies that prices generally go higher during holidays.
- When looking at weekly change, the number of avilable listings is the lowest on Friday and Saturday, indicating that there is a positive correlation between price and business at weekly basis.
- For monthly variability of available listing, while it shows the similar patterns during the summer season (Jun-Aug), the number of listing becomes the highest in December and drops to the lowest in Januanry. This might due to the closure of rent business during January. The number of available listings rapidly goes up during Feb and March, indicating those hosts might be opening up the house again to capture the increasing demand along the way from Spring to Summer.

### Data Preparation

```python
#Data preparation

#Removes rows missing price data since we are interested in price vairability
df_list.dropna(subset=['price'],inplace=True)

#Convert dollar dtype to float
df_list['price'] = df_list['price'].apply(lambda x:dollar_float(x))
df_list['monthly_price'] = df_list['monthly_price'].apply(lambda x:dollar_float(x))
df_list['weekly_price'] = df_list['weekly_price'].apply(lambda x:dollar_float(x))
df_list['security_deposit'] = df_list['security_deposit'].apply(lambda x:dollar_float(x))
df_list['cleaning_fee'] = df_list['cleaning_fee'].apply(lambda x:dollar_float(x))
df_list['extra_people'] = df_list['extra_people'].apply(lambda x:dollar_float(x))

#Convert percent to float
df_list['host_response_rate'] = df_list['host_response_rate'].apply(lambda x:perc_float(x))
```

```python
#Convert host_since to float days
df_list['host_since'] = (pd.to_datetime('2017-02-01') - pd.to_datetime(df_list['host_since'])) / np.timedelta64(1, 'D')

#By same token, we will convert last_scraped, first_review & last_review columns to float days
df_list['last_scraped'] = (pd.to_datetime('2017-02-01') - pd.to_datetime(df_list['last_scraped'])) / np.timedelta64(1, 'D')
df_list['first_review'] = (pd.to_datetime('2017-02-01') - pd.to_datetime(df_list['first_review'])) / np.timedelta64(1, 'D')
df_list['last_review'] = (pd.to_datetime('2017-02-01') - pd.to_datetime(df_list['last_review'])) / np.timedelta64(1, 'D')
```

### Handiling Categorical values

#### Handle Amenities

```python
df_list.amenities.head()
```

```python
all_ame = []
for x in df_list.amenities:
    all_ame.extend(x.replace('{','').replace('}','').replace('"','').split(','))
unique_ame = pd.unique(all_ame)
unique_ame
```

```python
#create matrix
zero_matrix = np.zeros((len(df_list),len(unique_ame)))
dummies_ame = pd.DataFrame(zero_matrix,columns=unique_ame)
```

```python
#fill the matrix
for i,ame in enumerate(df_list.amenities):
    indices = dummies_ame.columns.get_indexer(ame.replace('{','').replace('}','').replace('"','').split(','))
    dummies_ame.iloc[i,indices]=1
```

```python
dummies_ame
```

```python
sns.heatmap(dummies_ame.corr())
```

```python
df_list_ame_price = pd.concat([df_list['price'],dummies_ame],axis=1)
df_list_ame_price.corr()['price'].sort_values(ascending=False).plot.bar()
plt.xticks(rotation=90)
plt.show()
```

- Albeit weak, there are several items with correlation with price - Family/Kid Friendly, Cable TV/TV, Indoor Fireplace, and so forth.
- It is interesting that seemingly laxurary amenties such as Gym and Pool have weaker correlation with price than TV.

```python
df_list_ame_review = pd.concat([df_list['review_scores_rating'],dummies_ame],axis=1)
df_list_ame_review.corr()['review_scores_rating'].sort_values(ascending=False).plot.bar()
plt.xticks(rotation=90)
plt.show()
```

- Some amenities have postive correlation with positive correlation. However, given the weakness of correlation, it is more rational to consider that amenities are not determining factors for positive reveiw.

#### Cleaning Address Related Columns

```python
#Take a look at address related data
col_address = ['street','host_location', 'host_neighbourhood', 'neighbourhood_cleansed', 'neighbourhood', 
         'neighbourhood_group_cleansed','zipcode', 'city','state']
df_list[col_address].describe()
```

```python
for col in col_address:
    if col == 'street':
        continue
    elif col == 'host_location':
        continue
    else:
        print(df_list[col].value_counts())
```

- It seems 'neighbourhood_cleansed', 'neighbourhood_group_cleansed' and 'zipcode' covers most info for the other address related columns.
- Taking into the guest perspective, which will rely on Area name, rather than zipcode, 'neighbourhood_cleansed' and 'neighbourhood_group_cleansed' will be used for the following analysis.

```python
col_cat = ['neighbourhood_cleansed','neighbourhood_group_cleansed','calendar_updated','property_type',
          'smart_location','bed_type','host_response_time','cancellation_policy','room_type',
           'instant_bookable','is_location_exact','require_guest_profile_picture','host_identity_verified',
           'host_has_profile_pic','host_is_superhost','host_acceptance_rate','require_guest_phone_verification']
```

```python
#crate dummies columns and remove all categorical colums
for col in col_cat:
    df_list = pd.concat([df_list.drop(col,axis=1),pd.get_dummies(df_list[col],prefix=col,prefix_sep='_',drop_first=True, dummy_na=True)],axis=1)
```
